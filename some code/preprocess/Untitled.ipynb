```python
import numpy as np
import scipy.io as scio
import matplotlib.pyplot as plt
import matplotlib

goal1 =  scio.loadmat(r'D:\MCM_ICM\m_map\my_file\goals.mat')
y1 = goal1['goals'].reshape(500)[:150]
y2 = goal1['goals'].reshape(500)[:100]
```

```python
with plt.style.context('science'):
    fig,ax = plt.subplots(figsize=(4,4))
    x = np.linspace(1,150,150)
    font = {'weight' : 'bold',
        'size'   : 10}
    matplotlib.rc('font', **font)
    plt.plot(x,y1)
    
    ax.set_ylim([3000000, 11000000])
    #ax.set_xlim([0, 500])
    
    ax.set_xlabel(r"Time (year)",
              fontweight ='bold',
              #fontsize=18,
              loc="right")
    ax.set_ylabel(r"CSR (g/year)",
              fontweight ='bold',
              #fontsize=18,
              loc="top")
    x1 = np.linspace(0,100,100)
    plt.fill_between(x1,y2,edgecolor="b",alpha=.1)
    ax.legend(['Japan\'s CSV','CS'])
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    fig.savefig('ca5.pdf',dpi=400)
```
