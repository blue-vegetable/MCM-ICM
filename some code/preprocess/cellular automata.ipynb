```python
import numpy as np
import scipy.io as scio
import matplotlib.pyplot as plt
import matplotlib
goal1 =  scio.loadmat('D:\MCM_ICM\m_map\my_file\goals1.mat')
goal2 =  scio.loadmat('D:\MCM_ICM\m_map\my_file\goals2.mat')
goal3 =  scio.loadmat('D:\MCM_ICM\m_map\my_file\goals3.mat')
```

```python
np.linspace(1,500,500)
y1 = goal1['goals'].reshape(500)
y2 = goal2['goals'].reshape(500)
y3 = goal3['goals'].reshape(500)
```

```python
def linerline(x,a,b):
    return a*x+b
```

```python
with plt.style.context('science'):
    fig,ax = plt.subplots(figsize=(4,4))
    x = np.linspace(1,500,500)
    font = {'weight' : 'bold',
        'size'   : 10}
    matplotlib.rc('font', **font)
    plt.plot(x,y1,alpha=0.3)
    plt.plot(x,y2,alpha=0.3)
    plt.plot(x,y3,alpha=0.3)
    plt.plot(x,linerline(x,367,8.28e6),'--')
    plt.plot(x,linerline(x,439.7,8.559e6),'--')
    
    plt.plot(x,linerline(x,-220.4,8.55e6),'--')
   # plt.plot(x,linerline(x,445.8,8.889e6),'--')
    #plt.plot(x,linerline(x,109.7,8.878e6),'--')
    
    ax.set_ylim([3000000, 11000000])
    #ax.set_xlim([0, 500])
    ax.legend(['No Intervention','Ordinary Intervention','10 years longer','No Intervention','Ordinary Intervention','10 years longer'])
    ax.set_xlabel(r"Time (year)",
              fontweight ='bold',
              #fontsize=18,
              loc="right")
    ax.set_ylabel(r"CSR (g/year)",
              fontweight ='bold',
              #fontsize=18,
              loc="top")
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    fig.savefig('ca9.pdf',dpi=400)
```
